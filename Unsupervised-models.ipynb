```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate random dataset
np.random.seed(42)
data = np.random.randn(30, 2)

# Perform hierarchical clustering
Z = linkage(data, method='ward')

# Plot the dendrogram
plt.figure()
dendrogram(Z)
plt.xlabel('Data points')
plt.ylabel('Euclidean distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

# Cut the dendrogram at a certain height
max_d = 3
clusters = fcluster(Z, max_d, criterion='distance')

# Plot the clusters
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hierarchical Clustering Result')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate a dataset with 3 clusters
data, true_labels = make_blobs(n_samples=300, centers=3, random_state=42)

# Perform K-means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering Result')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
feature_names = iris.feature_names

# Standardize the dataset
scaler = StandardScaler()
data_std = scaler.fit_transform(data)

# Perform PCA with 2 components
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_std)

# Plot the first two principal components
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=target, cmap='viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Result')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster

# Sample dataset with fruit features (weight in grams and sweetness on a scale of 1-10)
data = pd.DataFrame({
    'Fruit': ['Apple', 'Orange', 'Banana', 'Strawberry', 'Blueberry', 'Grape', 'Pineapple', 'Watermelon', 'Mango', 'Lemon'],
    'Weight': [180, 160, 120, 15, 12, 4, 900, 5500, 400, 60],
    'Sweetness': [8, 6, 9, 10, 10, 10, 5, 7, 8, 3]
})

# Perform hierarchical clustering using the 'ward' method
Z = linkage(data[['Weight', 'Sweetness']], method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, labels=data['Fruit'].values, leaf_rotation=90)
plt.xlabel('Fruits')
plt.ylabel('Euclidean Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()

# Cut the dendrogram at a certain height to create clusters
max_d = 1500
clusters = fcluster(Z, max_d, criterion='distance')
data['Cluster'] = clusters

# Display the resulting clusters
print(data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Sample dataset with customer shopping habits (annual income and spending score)
data = pd.DataFrame({
    'Customer': ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10'],
    'Annual_Income': [40000, 60000, 80000, 100000, 120000, 140000, 160000, 180000, 200000, 220000],
    'Spending_Score': [10, 8, 6, 4, 2, 9, 7, 5, 3, 1]
})

# Standardize the dataset
scaler = StandardScaler()
data_std = scaler.fit_transform(data[['Annual_Income', 'Spending_Score']])

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 6):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(data_std)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 6), wcss)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method')
plt.show()

# Perform K-Means clustering with the optimal number of clusters (K = 3)
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(data_std)

# Plot the resulting clusters
plt.scatter(data['Annual_Income'], data['Spending_Score'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('K-Means Clustering Result')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['Species'] = iris.target

# Standardize the dataset
scaler = StandardScaler()
data_std = scaler.fit_transform(data.drop('Species', axis=1))

# Perform PCA with 2 principal components
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_std)

# Plot the first two principal components
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data['Species'], cmap='viridis', edgecolor='k', s=150)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Result')
plt.show()
```
